### Exploratory Data Analysis

* A **Probability Density Function** (PDF) is a mathematical function that describes the relative likelihood of a continuous random variable taking on a specific value. Unlike discrete variables, where exact outcomes have specific probabilities, continuous variables (like exact height or time) have an infinite number of possibilities, making the probability of any single exact value practically zero

* A **Box and whisker plots (boxplots)** use cases below
    * Outlier Detection & Treatment
    * Feature Distribution & Skewness
    * Evaluate cross-validation scores
* **Kurtosis** measures the "tailedness" of a data distribution, specifically indicating how much data resides in the extreme ends (outliers) versus the center

* A **Confidence Interval** (CI) is a range of values which is likely to contain (in repeated sampling) the true value of an unknown statistical parameter, such as a population mean. Rather than reporting a single point estimate

* **ks test**  used to check if a sample follows a specific distribution or if two samples come from the same distribution
